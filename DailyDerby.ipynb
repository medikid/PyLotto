```python
from dailyderby.models import iresult, idraw, idraw_c
from dailyderby.fetcher import Fetcher

class DailyDerby:
    DF_RESULTS = {}
    DF_RESULTS = {}
    
    def __init__(self):
        pass;
    
    def fetch_online_results(self, StartID=1234, StartDate='2020-07-10'):
        f = Fetcher();
        df_res = f.fetch_results(StartID, StartDate);
        
        return df_res;
        
    def derive_draws(self):
        results = iresult.iResult().db.session.query(iresult.iResult).all();
        for result in results:
            #print(result.draw_id, result.r);
            dr = idraw.iDraw(result.draw_id, [result.r1, result.r2, result.r3]);
            dr.derive();
            #print(dr.toString())
            dr.db_save()
            
    def derive_draws_c(self):
        results = iresult.iResult().db.session.query(iresult.iResult).all();
        for result in results:
            dr = idraw_c.iDrawC(result.draw_id, [result.r1, result.r2, result.r3]);
            dr.derive()
            dr.db_save()
            
    def get_draws_df(self):
        return idraw.iDraw().get_dataframe();
    
    def get_results_df(self):
        return iresult.iResult().get_dataframe();
    
    def get_master_df(self):
        pass;
```
